#### 1.Derive y explique la funci贸n de demanda de Consumo. Replique el gr谩fico siguiente:

##### a) Derivaci贸n de la funci贸n de demanda de consumo:
La funci贸n base de la demanda de Consumo es:

$$ C = C_0 + bYd  $$

siendo: 

$$T=YT$$

Entonces, al reemplazar valores, la funci贸n resultar铆a:

##### b) Explicaci贸n de la funci贸n

La funci贸n de demanda de consumo implica que las familias toman decisiones de gasto y ahorro. Adem谩s, se sabe que la suma de salarios de los trabajadores y de los beneficios de los propietarios del stock de capital es igual al Ingreso Agregado (Y) y parte de este ingreso (Y) se entrega al Estado como tributaci贸n (T), con lo que queda se decide consumir y ahorrar.

La funci贸n demanda de consumo: $C = C_0 + b(1-t)Y$, representa como parte del ingreso disponible se utiliza en los gastos de consumo. Adem谩s, en cuanto a los valores,$C_0$ es el nivel b谩sico de consumo, es decir, el consumo aut贸nomo, y $b$
representa la propensi贸n marginal($PMgc$) a consumir, es decir, el cambio en el consumo por unidad de cambio en el ingreso disponible (matem谩ticamente, se sabe que es la derivada de la funci贸n consumo con respecto al ingreso
disponible). Esta se encuentra entre cero y uno, 0 <  < 1. 

##### c) R茅plica de gr谩fico de funci贸n de demanda de Consumo:

#### 2.Derive y explique la funci贸n de Inversi贸n. Replique el gr谩fico siguiente:

##### a) Explicaci贸n de la funci贸n de demanda de Inversi贸n

Los gastos de inversi贸n son aquellos que aumentan el stock de capital y, por lo tanto, la producci贸n potencial o
producci贸n de largo plazo de la econom铆a. Estos pueden ser representados por medio de la funci贸n de Demanda de Inversi贸n: 

$$I= I_0-hr$$

donde $I_0$ representa la inversi贸n aut贸noma (valor fijo),$h$ representa la sensibilidad de un inversionista ante la tasa de inter茅s real y $r$ representa la tasa de inter茅s real. 

Adem谩s, los valores de $r$ y $h$ se encuentran en un rango de 0 a 1.Entonces, hay una relaci贸n inversa entre la tasa de inter茅s real $r$ y la funci贸n de demanda de Inversi贸n $I$ y la inversi贸n aut贸noma $I_0$ es el componente aut贸nomo que depende b谩sicamente de las expectativas de los inversionistas.

##### b)Gr谩fico de la funci贸n de Inversi贸n:

```python
#Par谩metros
Io = 40
h = 0.7

r = np.arange(start = 0.01, stop = 1, step =0.01)  
r

# Ecuaci贸n de la curva del ingreso de equilibrio

def I(Io, h, r):
    I = Io - h*r
    return I

I = I(Io, h, r)
```

##### a) Encuentre el nivel de Ingreso de equilibrio Ye

Para determinar el ingreso de equilibrio ($Y^e$), el ingreso agregado ($Y$) debe ser igual a la demanda agregada($DA$), es decir, $Y=DA$. As铆, a partir de esta condici贸n de equilibrio se obtiene la ecuaci贸n del ingreso de equilibrio ($Y^e$) a corto plazo.

Entonces, definimos la ecuaci贸n de la Demanda Agregada:

$$ DA = (C_0 + bYd + I_0 + G_0 + X_0 - hr) - mYd $$

Luego, al reemplazar las ecuaciones del modelo (consumo, inversion, exportaciones, importaciones, gasto del gobierno e ingreso disponible) y realizar algunas operaciones algebraicas obtenemos:

$$ DA = (C_0 + b(1-t)Y + I_0 + G_0 + X_0 - hr) - m(1-t)Y $$

$$ DA = (C_0 + I_0 + G_0 + X_0 - hr) + (b - m)(1 - t)Y $$

De forma corta:

$$ DA = 伪_0 + 伪_1Y $$

Donde $ 伪_0 = (C_0 + I_0 + G_0 + X_0 -hr)$ es el intercepto y $ 伪_1 = (b - m)(1 - t) $ es la pendiente de la funci贸n

Ahora, considerando la condici贸n de equilibrio $Y = DA$, la ecuaci贸n del ingreso de equilibrio es:

$$ Y = (C_0 + I_0 + G_0 + X_0 - hr) + (b - m)*(1 - t)Y $$

$$ Y =  \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

##### b) Analice una Pol铆tica fiscal expansiva con una reducci贸n de la Tasa de Tributaci贸n ($t < 0$). El an谩lisis debe ser intuitivo, matem谩tico y gr谩fico.
- An谩lisis intuitivo: 驴Hay contradicci贸n?

$$ t  Co  DA  DA > Y  Y $$

$$ t  M  DA  DA < Y  Y $$

- An谩lisis matem谩tico: $t < 0    驴Y?$

```python
Co, Io, Go, Xo, h, r, b, m, t = symbols('Co Io Go Xo h r b m t')

f = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))

df_t = diff(f, t)
df_t #Y/t
```

Considernado el diferencial de $t$:

$$ \frac{Y}{t} = \frac{(m-b)(Co + Go + Io + Xo - hr)}{(1-(1-t)(b-m)+1)^2} $$

- Sabiendo que b > m, entonces $(m-b) < 0$
- Los componentes aut贸nomos no cambian: $C_0 = I_0 = X_0 = h = r = 0$
- Cualquier n煤mero elevado al cuadrado ser谩 positivo: $ (1-(1-t)(b-m)+1)^2 > 0 $

Entonces:

$$ \frac{Y}{t} = \frac{(-)}{(+)} $$

Dado que $t < 0$, la divisi贸n de dos positivos da otro positivo:

$$ \frac{Y}{(-)} = \frac{(-)}{(+)} $$

$$ Y = \frac{(-)(-)}{(+)} $$
